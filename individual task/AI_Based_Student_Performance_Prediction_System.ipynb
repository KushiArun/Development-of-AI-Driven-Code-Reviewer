```python
import sqlite3
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
data = {
    "attendance": [90, 60, 75, 40, 85, 55, 95, 65],
    "internal_marks": [80, 45, 60, 35, 78, 50, 88, 55],
    "assignment_score": [85, 40, 70, 30, 80, 45, 90, 60],
    "performance": ["Good", "Pass", "Pass", "Fail", "Good", "Pass", "Good", "Pass"]
}

df = pd.DataFrame(data)
df
```

```python
X = df[["attendance", "internal_marks", "assignment_score"]]
y = df["performance"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

print("Model trained successfully")
```

```python
new_student = [[70, 65, 75]]  # attendance, marks, assignment
prediction = model.predict(new_student)

print("Predicted Performance:", prediction[0])
```

```python
conn = sqlite3.connect("student_performance.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS predictions (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    attendance INTEGER,
    internal_marks INTEGER,
    assignment_score INTEGER,
    predicted_performance TEXT
)
""")

conn.commit()
print("Database created")
```

```python
cursor.execute("""
INSERT INTO predictions (attendance, internal_marks, assignment_score, predicted_performance)
VALUES (?, ?, ?, ?)
""", (70, 65, 75, prediction[0]))

conn.commit()
print("Prediction stored in database")
```
